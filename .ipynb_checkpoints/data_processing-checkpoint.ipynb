```python
df = pd.read_csv('43563d52-ce16-4d34-bbd4-e8ce892fa9b7.csv')
```

```python
def process(sample_data, dtype):
    s_df = pd.DataFrame(eval(sample_data))
    if s_df.empty:
        return pd.DataFrame()
    aggreagted = s_df.groupby('price').sum()
    if dtype == 'ask':
        return aggreagted.sort_index()
    elif dtype == 'bid':
        return aggreagted.sort_index(ascending=False)
    else:
        raise ValueError()
```

```python
def extract_data(file_path):
    df = pd.read_csv(file_path)
    
    res = df[['created', 'pair']].rename(columns={'created': 'trade_time'})
    new_columns = [f'bid{i}' for i in range(1, 21)]
    new_columns += [f'bid{i}_vol' for i in range(1, 21)]
    new_columns += [f'ask{i}' for i in range(1, 21)]
    new_columns += [f'ask{i}_vol' for i in range(1, 21)]
    res = res.reindex(columns=res.columns.tolist() + new_columns)
    
    for row_number in range(len(df)):
        bid_data = df.loc[row_number, 'bids']
        process_data = process(bid_data, dtype='bid')
        for i in range(min(20, len(process_data))):
            res.loc[row_number, [f'bid{i+1}', f'bid{i+1}_vol']] = [process_data.index[i], process_data.values[i]]

        ask_data = df.loc[row_number, 'asks']
        process_data = process(ask_data, dtype='ask')
        for i in range(min(19, len(process_data))):
            res.loc[row_number, [f'ask{i+1}', f'ask{i+1}_vol']] = [process_data.index[i], process_data.values[i]]
    return res
```

```python
%%time
extract_data('43563d52-ce16-4d34-bbd4-e8ce892fa9b7.csv')
```
